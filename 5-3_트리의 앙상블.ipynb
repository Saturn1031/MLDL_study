```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

wine = pd.read_csv('https://bit.ly/wine_csv_data')

data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)     # 랜덤 포레스트 (랜덤 훈련 세트 사용)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)  # 교차 검증

print(np.mean(scores['train_score']), np.mean(scores['test_score']))    # 과대적합
```

```python
rf.fit(train_input, train_target)
print(rf.feature_importances_)      # 특성 중요도
```

```python
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)     # 부트스트랩 샘플에 포함되지 않는 OOB 샘플로 검증

rf.fit(train_input, train_target)
print(rf.oob_score_)    # OOB 점수
```

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)   # 엑스트라 트리 (전체 훈련 세트 사용)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))

et.fit(train_input, train_target)
print(et.feature_importances_)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)    # 그레이디언트 부스팅 (깊이가 3인 결정 트리 100개 사용)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)   # 결정 트리 개수 500개
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))

gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)   # 히스토그램 기반 그레이디언트 부스팅 (특성을 256개 구간으로 나누어 최적의 분할을 찾음)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10, random_state=42, n_jobs=-1)

print(result.importances_mean)  # 특성 중요도 평균
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10, random_state=42, n_jobs=-1)

print(result.importances_mean)
```

```python
hgb.score(test_input, test_target)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(tree_method='hist', random_state=42)    # 대표적인 그레이디언트 부스팅 라이브러리
scores = cross_validate(xgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier(random_state=42)   # 마이크로소프트 개발 히스토그램 기반 그레이디언트 부스팅 라이브러리
scores = cross_validate(lgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```
